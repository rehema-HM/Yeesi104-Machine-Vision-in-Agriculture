Furthermore the imread method can also read the image as a color image or grayscale by specifying the flag

Read an image
cv2.imread() Second argument is a flag which specifies the way image should be read.

flag  integer value  description

cv2.IMREAD_COLOR  1  Loads a color image.

cv2.IMREAD_GRAYSCALE  0  Loads image in grayscale mode

cv2.IMREAD_UNCHANGED  -1  Loads image as such including alpha channel

(The alpha channel is a special channel that handles transparency. When an image has an alpha channel on it, it means you can adjust the image's opacity levels and make bits translucent or totally see-through. The alpha channel is instrumental when you want to remove the background from an image)


````python
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/yeesi/images_yeesi/American-Fall-Army-Worm- trageddy.jpeg',1)
cv2_imshow(img)
````

````python
#Convert the image to grayscale using:
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
````

**Saving the Image**

The image will be saved in the current working directory. The first parameter corresponds to the name of the file in which the image is to be saved and the second parameter is the variable that contains the image (pixel information).

````python
#Saving the Image
cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/yeesi/images_yeesi/Mandrill_grey.jpg', gray)
````
